# Time series DataSet Analysis

- Define the DataSet Dir

```python
DataSet_dir = './DataSet'
TrainingSet = '{}/DailyDelhiClimateTrain.csv'.format(DataSet_dir)
TestSet = '{}/DailyDelhiClimateTest.csv'.format(DataSet_dir)
```

- Import pkgs

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv(TrainingSet)
df_test = pd.read_csv(TestSet)
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
df_train['date'] = pd.to_datetime(df_train['date'])
df_train.info()
```

```python
df_test.head()
df_test['date'] = pd.to_datetime(df_test['date'])
df_test.info()
```

- Explore Data

```python
plt.plot(df_train['date'], df_train['meantemp'])
plt.title('Date/MeanTemp Plot')
plt.xlabel('Date')
plt.ylabel('MeanTemp')
plt.xticks(rotation=90)
```

```python
i = (len(df_train.columns) - 1) * 100 + 10
for col in df_train.columns:
    if col == 'date':
        continue
    plt.subplot(111)
    plt.plot(df_train['date'], df_train[col])
    plt.title('Date/{} Plot'.format(col.title()))
    plt.xlabel('Date')
    plt.ylabel(col.title())
    plt.xticks(rotation=90)
    i += 10
    plt.show()
```
